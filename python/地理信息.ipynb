```python
import folium
import pandas as pd

# define the world map
world_map = folium.Map()

# display world map
world_map
```

```python
latitude = 15.196337
longitude = 104.848005
```

```python
# Create map and display it
san_map = folium.Map(location=[latitude, longitude], zoom_start=12)
# Display the map 
san_map
```

```python
cdata=pd.read_csv(r"D:\Onedrive\68，地理信息\dcsp.csv")
cdata.head()
```

```python
# get the first 200 crimes in the cdata
data = cdata

# Instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# Loop through the 200 crimes and add each to the incidents feature group
for lat, lng, in zip(cdata.lat, cdata.lng):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.4
        )
    )

# Add incidents to map
san_map = folium.Map(location=[latitude, longitude], zoom_start=12)
san_map.add_child(incidents)
```

```python
data2=pd.read_csv(r"D:\Onedrive\68，地理信息\15hub.csv")
# add pop-up text to each marker on the map
latitudes = list(data2.lat)
longitudes = list(data2.lng)
labels = list(data2.name)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(san_map)    
    
# add incidents to map
san_map.add_child(incidents)
```

```python
data3=pd.read_csv(r"D:\Chrome下载\sqlresult_4390266.csv")
# add pop-up text to each marker on the map
latitudes = list(data3.lat)
longitudes = list(data3.lng)
labels = list(data3.store_name)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label,icon=folium.Icon(color='red', icon='info-sign')).add_to(san_map)    
# add incidents to map

san_map.add_child(incidents)
```
